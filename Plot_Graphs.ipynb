# Context Retrieval - FQuad
## TW-IDF with AND policy with words

Ce modèle calcule les contextes les plus pertinents, étant donné une question.

```python
top_k_threshold = []
context_retrieval_score = []
article_retrieval_score = []

with open("fquad_retrieval_and_policy_tw_idf.txt", "r") as result_file:
    lines = result_file.readlines()
    for line in lines:
        k, context_score, article_score = line.rstrip('\n').split(' ')
        top_k_threshold.append(int(k))
        context_retrieval_score.append(float(context_score)*100)
        article_retrieval_score.append(float(article_score)*100)
```

```python
plt.figure(figsize=(15, 10))
plt.ylim(0, 100)
plt.plot(top_k_threshold, context_retrieval_score, label="Context Retrieval Score")
plt.plot(top_k_threshold, article_retrieval_score, label="Article Retrieval Score")
plt.xticks(fontsize=25)
plt.xlabel("Top-k threshold", fontsize=20)
plt.yticks(fontsize=25)
plt.ylabel("Accuracy in %", fontsize=20)
plt.title("Context and Article Retrieval Score with TW-IDF model - AND Logic / 4900 Contexts", fontsize=25, y=-0.18)
plt.legend(fontsize=14)
plt.savefig("result_tw_idf_context_article_score_and_logic.png")
plt.show()
```
